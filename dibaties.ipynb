```python
import pandas as pd
```

```python
df = pd.read_csv('framingham.csv')
```

```python
df.isnull()
```

```python
df.info()
```

```python
#drop null value
df.dropna(inplace =True)
```

```python
df.info()
```

```python
#preparing data
x = df.drop('TenYearCHD', axis = 'columns')
```

```python
y = df['TenYearCHD']
```

```python
#import 
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test, y_train,y_test = train_test_split(x,y, train_size =0.65)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
#model trainng
model.fit(x_train, y_train)
```

```python
# predicting data for test data
y_pred = model.predict(x_test)
```

```python
#predicated data
y_pred
```

```python
#problity
model.predict_proba(x_test)
```

```python
#model score
model.score(x_test,y_test)
```

```python
#predicted & origanal valure
df1 = pd.DataFrame({'original':y_test, 'predicted': y_pred})
```

```python
df1.to_csv('dia.csv')
```
